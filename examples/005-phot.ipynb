```python
site = Coordinates.location(45, 45, 2000)
site
```

```python
sources = Fixer.list_to_source(
    [
        [1140.280524, 1482.407145],
        [899.134273, 973.359417]
    ]
)
sources
```

```python
ape = 10
ann = 15
dan = 25
```

```python
phot_sep = aphot.sep(sources, ape, extract=["JD", "FILTER", "AIRMASS"])
phot_sep
```

```python
phot_pht = aphot.photutils(sources, ape, ann, extract=["JD", "FILTER", "AIRMASS"])
phot_pht
```

```python
phot_irf = aphot.iraf(sources, ape, ann, dan, extract=["JD", "FILTER", "AIRMASS"])
phot_irf
```
